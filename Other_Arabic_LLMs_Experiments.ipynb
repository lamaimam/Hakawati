#1. Llama 3.0 - 8B LLM Model

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

# Load model
model_name = "meta-llama/Meta-Llama-3-8B"

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load the model and move it to the GPU manually
model = AutoModelForCausalLM.from_pretrained(model_name)

# Check if GPU is available, and move the model to GPU
if torch.cuda.is_available():
    model = model.to("cuda")
    print("Model loaded on GPU.")
else:
    print("GPU not available, model loaded on CPU.")

# Create a text generation pipeline
text_generation_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    pad_token_id=tokenizer.eos_token_id,
    truncation=True,
    device=0 if torch.cuda.is_available() else -1  # Ensure correct device is used in the pipeline
)

# Generate a story
def get_llama3_output(prompt: str) -> None:
    """
    Generate a response from the Llama model.

    Parameters:
        prompt (str): The user's input/question for the model.

    Returns:
        None: Prints the model's response.
    """
    sequences = text_generation_pipeline(
        prompt,
        do_sample=True,
        top_k=50,
        top_p=0.95,
        temperature=0.7,
        num_return_sequences=1,
        max_length=256
    )
    print("LLaMa3 answer:", sequences[0]['generated_text'])
```

## 1.1 Different Prompts Using Llama 3.1 8B Model

### 1.1.1 Short English prompt.



```python
prompt = "Write a story for kids"

get_llama3_output(prompt)
```

### 1.1.2 Short Arabic prompt (Instruct)



```python
prompt = "أكتب قصة قصيرة تكون مفهومة للأطفال "
get_llama3_output(prompt)
```

### 1.1.3 Short Arabic prompt that starts the sentence for the model.

```python
prompt = "كان هناك طفل اسمه موسى"
get_llama3_output(prompt)
```

```python
from transformers import AutoModelForCausalLM, pipeline, GPT2TokenizerFast
from arabert.preprocess import ArabertPreprocessor

MODEL_NAME='aubmindlab/aragpt2-mega'
arabert_prep = ArabertPreprocessor(model_name=MODEL_NAME)

# Testing a basic short arabic prompt
text="اكتب قصة قصيرة"
text_clean = arabert_prep.preprocess(text)

# Load the model
model = AutoModelForCausalLM.from_pretrained(MODEL_NAME, trust_remote_code=True)
tokenizer = GPT2TokenizerFast.from_pretrained(MODEL_NAME)
generation_pipeline = pipeline(
    "text-generation", model=MODEL_NAME, trust_remote_code=True
)

generation_pipeline(text,
    pad_token_id=generation_pipeline.tokenizer.eos_token_id,
    num_beams=10,
    max_length=200,
    top_p=0.9,
    repetition_penalty = 3.0,
    no_repeat_ngram_size = 3)[0]['generated_text']
```
