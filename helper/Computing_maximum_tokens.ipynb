### This code is to decide the value for maximum tokens
(based on calculating the maximum tokens of a prompt-story pair in our dataset

```python
import pandas as pd
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("unsloth/Meta-Llama-3.1-8B-Instruct-bnb-4bit")

df = pd.read_excel('/content/merged_file.xlsx')

# Extract Prompts and Stories and convert them to list
prompts = df['Prompt'].tolist()
stories = df['Story'].tolist()

# Combines a prompt and story into a single string.
def combine_texts(prompt, story, eos_token):
  # delimiters: ###before the prompt and @@@ between the prompt and the story.
  return f"###{prompt}@@@{story}{eos_token}"

# combining each prompt with its story to calculate the maximun token of a prompt-story pair
max_combined_tokens = max(
    len(tokenizer.encode(combine_texts(prompt, story, tokenizer.eos_token), add_special_tokens=False))
    for prompt, story in zip(prompts, stories)
)

print(f"Maximum number of tokens in the combined prompt and story: {max_combined_tokens}")
```
